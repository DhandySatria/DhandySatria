```python
def IniTokkoKhopie():
    from IPython.display import clear_output
    print("List Harga Bijie Khopie:")    
    print("01 = Gayo")  
    print("02 = Kintamani") 
    print("03 = Bajawa") 
    print("04 = Garut")
    print("05 = Semendo") 
    print("06 = Kerinci")
    print("07 = Alat Cafe Press")
    print("08 = Alat Coffee Grinders")
    print("09 = Alat Bean Plus")
    print("10 = Alat Aero Press")               



    jumlahtotal = 0     
    hargatotal = 0          
    HargaGayo = (170000)         
    HargaKintamani = (150000)
    HargaBajawa = (150000)
    HargaGarut = (190000)
    HargaSemendo = (180000)
    HargaKerinci = (150000)
    HargaCafePress = (410000)
    HargaCoffeeGrinders = (1424000)
    HargaBeanPlus = (2450000)
    HargaAeroPress = (19000)
    barang = 1                    
    
    while barang != 0 :              
        KodeBarang = input("Masukkan kode barang : ")
        if KodeBarang == "01":                           
            print ("Aceh Gayo Honey = " + str(HargaGayo))  
            JumlahA = int(input("Jumlah Barang : "))    
            hargaA = HargaGayo * JumlahA            
            print (hargaA)                              
            jumlahtotal = jumlahtotal + JumlahA         
            hargatotal = hargatotal + hargaA            
        elif KodeBarang == "02":    
            print ("Kintamani = " + str(HargaKintamani))        
            JumlahB =int(input("Jumlah Barang : "))
            hargaB = HargaKintamani * JumlahB
            print (hargaB)
            jumlahtotal = jumlahtotal + JumlahB
            hargatotal = hargatotal + hargaB
        elif KodeBarang == "03":
            print ("Bajawa = " + str(HargaBajawa))
            JumlahC =int(input("Jumlah Barang : "))
            hargaC = HargaBajawa * JumlahC
            print (hargaC)
            jumlahtotal = jumlahtotal + JumlahC
            hargatotal = hargatotal + hargaC
        elif KodeBarang == "04":
            print ("Garut = " + str(HargaGarut))
            JumlahD =int(input("Jumlah Barang : "))
            hargaD = HargaGarut * JumlahD
            print (hargaD)
            jumlahtotal = jumlahtotal + JumlahD
            hargatotal = hargatotal + hargaD
        elif KodeBarang == "05":
            print ("Semendo = " + str(HargaSemendo))
            JumlahE =int(input("Jumlah Barang : "))
            hargaE = HargaSemendo * JumlahE
            print (hargaE)
            jumlahtotal = jumlahtotal + JumlahE
            hargatotal = hargatotal + hargaE
        elif KodeBarang == "06":
            print ("Kerinci = " + str(HargaKerinci))
            JumlahF =int(input("Jumlah Barang : "))
            hargaF = HargaKerinci * JumlahF
            print (hargaF)
            jumlahtotal = jumlahtotal + JumlahF
            hargatotal = hargatotal + hargaF
        elif KodeBarang == "07":
            print ("Alat Cafe Press = " + str(HargaCafePress))
            JumlahG =int(input("Jumlah Barang : "))
            hargaG = HargaCafePress * JumlahG
            print (hargaG)
            jumlahtotal = jumlahtotal + JumlahG
            hargatotal = hargatotal + hargaG
        elif KodeBarang == "08":
            print ("Alat Coffee Grinders = " + str(HargaCoffeeGrinders))
            JumlahH =int(input("Jumlah Barang : "))
            hargaH = HargaCoffeeGrinders * JumlahH
            print (hargaH)
            jumlahtotal = jumlahtotal + JumlahH
            hargatotal = hargatotal + hargaH
        elif KodeBarang == "09":
            print ("Alat Bean Plus = " + str(HargaBeanPlus))
            JumlahI =int(input("Jumlah Barang : "))
            hargaI = HargaBeanPlus * JumlahI
            print (hargaI)
            jumlahtotal = jumlahtotal + JumlahI
            hargatotal = hargatotal + hargaI
        elif KodeBarang == "10":
            print ("Alat Aero Press = " + str(HargaAeroPress))
            JumlahJ =int(input("Jumlah Barang : "))
            hargaJ = HargaAeroPress * JumlahJ
            print (hargaJ)
            jumlahtotal = jumlahtotal + JumlahJ
            hargatotal = hargatotal + hargaJ
        elif KodeBarang == "selesai":                     
            print("Total Item = " + str(jumlahtotal))     
            print("Total Bayar = " + str(hargatotal))     
            if (hargatotal>1000000):                        
                print("anda mendapatkan diskon sebesar Rp", hargatotal*0.1)
                hargasetelahdiskon = hargatotal-hargatotal*0.1
                print("Harga total setelah diskon anda menjadi Rp", hargasetelahdiskon)
            else:
                print("anda tidak mendapatkan diskon")    
                hargasetelahdiskon = hargatotal*1

            bayar = int(input("jumlah uang = "))   
            if (bayar>=hargasetelahdiskon):  
                print("Jumlah kembalian anda adalah Rp", bayar-hargasetelahdiskon)
                print(" ")
                print("Terimakasih telah berbelanja di IniTokkoKhopie, Hitam tak selalu kotor, Pahit tak harus pedih :)")
                if input("press space to restart: ") != "space":  
                    clear_output()                               
                    IniTokkoKhopie()
                    
            else:
                print ("uang anda kurang untuk melakukan transaksi ini") 
                print(" ")
                print("Total Item = " + str(jumlahtotal))
                print("Total Bayar = " + str(hargasetelahdiskon))  
                bayar = int(input("jumlah uang = "))
                if (bayar>=hargasetelahdiskon):           
                    print("Jumlah kembalian anda adalah Rp", bayar-hargasetelahdiskon) 
                    print(" ")
                    print("Terimakasih telah berbelanja di IniTokkoKhopie, Hitam tak selalu kotor, Pahit tak harus pedih :)")                        
                    if input("press space to restart: ") != "space":
                        clear_output()
                        IniTokkoKhopie()
                    
                else:
                    print("mohon maaf, uang anda masih belum cukup untuk pembelian produk ini, Silakan kembali lagi lain waktu. Terimakasih!") 
                    if input("press space to restart: ") != "space": 
                        clear_output()
                        IniTokkoKhopie() 
    
```
